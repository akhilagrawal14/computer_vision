```python
#importing required libraries
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt
```

```python
# reading the image
img = imread(r"C:\Users\akhil\OneDrive\Pictures\cv_input_output\input_1.jpg")
plt.axis("off")
plt.imshow(img)
print(img.shape)
```

```python
ratio=img.shape[1]/img.shape[0]  #width/height

fixed_resized_height=256

resized_shape=(fixed_resized_height,int(ratio*fixed_resized_height),3)    #1280/448= ?/256
```

```python
# resizing image
resized_img = resize(img, resized_shape)
plt.axis("off")
plt.imshow(resized_img)
print(resized_img.shape)
```

```python
#creating hog features
fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(16, 16),
                    cells_per_block=(2, 2), visualize=True, multichannel=True,feature_vector = False)
plt.axis("off")
plt.imshow(hog_image, cmap="gray")
```
