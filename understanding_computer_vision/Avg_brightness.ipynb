```python
import cv2
import numpy as np
import os

import matplotlib.pyplot as plt
```

```python
img1=cv2.imread(r"C:\Users\akhil\Downloads\IMG_20221010_185358.jpg")
```

```python
def average_brighness(rgb_image):
    hsv=cv2.cvtColor(rgb_image,cv2.COLOR_RGB2HSV)
    value_channel=hsv[:,:,2]
    sum_brightness=np.sum(value_channel)
    
    h,w=value_channel.shape
    area=h*w
    
    avg_brightness=sum_brightness/area
    
    return avg_brightness
```

```python
rgb_image=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

print(average_brighness(rgb_image))

plt.imshow(rgb_image)

plt.show()
```
