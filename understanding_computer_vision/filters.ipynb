```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread(r"D:\singapore_pics\by me\IMG_20221013_191616.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)


plt.imshow(img)
plt.show()
```

```python
#1st convert image to grayscale
gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
#sobel x filter
sobel_x=np.array([[-1,0,1],
                 [-2,0,2],
                 [-1,0,1]])

filtered_image=cv2.filter2D(gray,-1,sobel_x)

plt.imshow(filtered_image,cmap="gray")
plt.show()
```

```python
#1st convert image to grayscale
gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
#sobel x filter
sobel_y=np.array([[-1,-2,-1],
                 [0,0,0],
                 [1,2,1]])

filtered_image2=cv2.filter2D(gray,-1,sobel_y)

plt.imshow(filtered_image2,cmap="gray")
plt.show()
```
